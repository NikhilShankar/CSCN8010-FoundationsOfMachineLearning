### **4. Modeling**

##### **4.1 Splitting dataset into training and test dataset**

```python
# In order to split data frame into two sets 
# training and test dataset we need to import scikit learning library 'model_selection' module.
# We can just split using array split options also but then it won't be randomized. The usage of the following function from scikit 
# allows to specify a random_state parameter which promises reproducibility even when the split is randomized. ie. If same dataset is passed to the train_test_split 
# function with the same parameter values and random_state key then the split would always be identical.

from sklearn.model_selection import train_test_split

training_df, test_df = train_test_split(housing_df_processed, test_size=0.2, random_state=25)
print(f'Training set count : {len(training_df)}')
print(f'Test set count : {len(test_df)}')
```

#### **4.2 Creating Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression

# Linear regression model for all three independent variable
multivariateParams = ['median_income', 'population', 'households']

lrm_median_income = LinearRegression()
lrm_median_income.fit(training_df[multivariateParams], training_df[['median_house_value']])

lrm_population = LinearRegression()
lrm_population.fit(training_df[['population']], training_df[['median_house_value']])

lrm_households = LinearRegression()
lrm_households.fit(training_df[['households']], training_df[['median_house_value']])
```

#### **4.3 Intercepts and Error for the three models**

##### **Training Set**

```python
#importing mean_squared_error and mean_absolute_error from scikit learn library
from sklearn.metrics import mean_squared_error, mean_absolute_error

intercept_lrm_median_income = lrm_median_income.intercept_
coeff_lrm_median_income = lrm_median_income.coef_
mse_lrm_median_income = mean_squared_error(training_df[['median_house_value']], lrm_median_income.predict(training_df[multivariateParams]))
mae_lrm_median_income = mean_absolute_error(training_df[['median_house_value']], lrm_median_income.predict(training_df[multivariateParams]))

intercept_lrm_population = lrm_population.intercept_
coef_lrm_population = lrm_population.coef_
mse_lrm_population = mean_squared_error(training_df[['median_house_value']], lrm_population.predict(training_df[['population']]))
mae_lrm_population = mean_absolute_error(training_df[['median_house_value']], lrm_population.predict(training_df[['population']]))

intercept_lrm_households = lrm_households.intercept_
coef_lrm_households = lrm_households.coef_
mse_lrm_households = mean_squared_error(training_df[['median_house_value']], lrm_households.predict(training_df[['households']]))
mae_lrm_households = mean_absolute_error(training_df[['median_house_value']], lrm_households.predict(training_df[['households']]))

model_data = {
    'Parameter': ['Median Income', 'Population', 'Number of Households', ],
    'Intercept': [intercept_lrm_median_income, intercept_lrm_population, intercept_lrm_households],
    'Coefficient': [coeff_lrm_median_income, coef_lrm_population, coef_lrm_households],
    'Mean Squared Error': [mse_lrm_median_income, mse_lrm_population, mse_lrm_households],
    'Mean Absolute Error': [mae_lrm_median_income, mae_lrm_population, mae_lrm_households],
}

model_df = pd.DataFrame(model_data)
```

##### **Table of Values - Training Set** 

```python
display(model_df)
mean_of_house_value = training_df['median_house_value'].mean()
print(f'The mean value of median_house_value is {mean_of_house_value}')
```

##### **Test Set**

```python
#importing mean_squared_error and mean_absolute_error from scikit learn library
from sklearn.metrics import mean_squared_error, mean_absolute_error

mse_lrm_median_income_test = mean_squared_error(test_df[['median_house_value']], lrm_median_income.predict(test_df[multivariateParams]))
mae_lrm_median_income_test = mean_absolute_error(test_df[['median_house_value']], lrm_median_income.predict(test_df[multivariateParams]))

mse_lrm_population_test = mean_squared_error(test_df[['median_house_value']], lrm_population.predict(test_df[['population']]))
mae_lrm_population_test = mean_absolute_error(test_df[['median_house_value']], lrm_population.predict(test_df[['population']]))

mse_lrm_households_test = mean_squared_error(test_df[['median_house_value']], lrm_households.predict(test_df[['households']]))
mae_lrm_households_test = mean_absolute_error(test_df[['median_house_value']], lrm_households.predict(test_df[['households']]))

model_data_test = {
    'Parameter': ['Median Income', 'Population', 'Number of Households', ],
    'Mean Squared Error': [mse_lrm_median_income_test, mse_lrm_population_test, mse_lrm_households_test],
    'Mean Absolute Error': [mae_lrm_median_income_test, mae_lrm_population_test, mae_lrm_households_test],
}

model_df_test = pd.DataFrame(model_data_test)
```

##### **Table of Values - Test Set** 

```python
display(model_df_test)
mean_of_house_value = test_df['median_house_value'].mean()
print(f'The mean value of median_house_value is {mean_of_house_value}')
```

### **5. Evaluation**

#### **5.1 Plots**

```python
import numpy as np
import warnings
import math

def scatter_plot_with_regression_line(testdf, trainingdf, lrm, x_column_name, 
                                      y_column_name, title_test, title_training, 
                                      xlabel, ylabel, mse_test, mae_test, mse_training, mae_training):
    warnings.filterwarnings('ignore')
    fig, ax = plt.subplots(1, 2, figsize=(10,5))
    sns.scatterplot(x=x_column_name, y=y_column_name, data=test_df, color="red", ax=ax[0])
    ax[0].set_title(title_test)
    ax[0].set_xlabel(xlabel)
    ax[0].set_ylabel(ylabel)
    ax[0].text(x=5, y=10, s=f"Intercept : {lrm.intercept_}\nCoefficient : {lrm.coef_} \nMSE Test : {mse_test}\nMAE Test: {mae_test}")
    ax[1].text(x=5, y=10, s=f"\nMSE Training : {mse_training}\nMAE Training: {mae_training}")


    sns.scatterplot(x=x_column_name, y=y_column_name, data=training_df, color="blue", ax=ax[1])
    ax[1].set_title(title_training)
    ax[1].set_xlabel(xlabel)
    ax[1].set_ylabel(ylabel)



    #Creating a np array and then converting into dataframe for use in sns line plot function.
    # This had to be done since plt line plot was not working with two columns in a single row for plot graph
    actual_test_values = test_df[[x_column_name]]
    predicted_test_house_values = lrm.predict(actual_test_values)
    np_predicted_test = np.hstack((actual_test_values, predicted_test_house_values))
    df_np_predicted_test = pd.DataFrame(np_predicted_test, columns=[xlabel, ylabel])

    actual_training_values = training_df[[x_column_name]]
    predicted_training_house_values = lrm.predict(actual_training_values)
    np_predicted_training = np.hstack((actual_training_values, predicted_training_house_values))
    df_np_predicted_training = pd.DataFrame(np_predicted_training, columns=[xlabel, ylabel])

    sns.lineplot(data=df_np_predicted_test, x=xlabel, y=ylabel, color="black", ax=ax[0])
    sns.lineplot(data=df_np_predicted_training, x=xlabel, y=ylabel, color="black", ax=ax[1])
    plt.tight_layout()
    plt.show()
    return
```

#### **5.2 Analysis**

##### **5.2.1 Error Ratio**
Calculating error ratio from mean absolute error and mean value
> Error ratio is calculated by dividing Mean Absolute Error by Mean Value of the parameter.

```python
from math import atan, degrees
from sklearn.metrics import r2_score

error_ratio = pd.DataFrame({'ErrorRatio': [(mae_lrm_median_income_test/mean_of_house_value)*100, (mae_lrm_population_test/mean_of_house_value)*100, (mae_lrm_households_test/mean_of_house_value)*100]})
```

##### **5.2.2 RMSE - Root Mean Squared Error**
RMSE is the root of Mean Squared Error. This is a good indicator since MSE is denoted by 'square' of the variable. RMSE solves this by taking the root of MSE and thereby converting it into the same unit as the independent variable.

```python
rmse = pd.DataFrame({'RMSE': [(np.sqrt(mse_lrm_median_income_test)), np.sqrt(mse_lrm_population_test), np.sqrt(mse_lrm_households_test)]})
```

##### **5.2.3 R-Squared**
rsquared is a value that denotes the variance of housing value based on each parameter.
- value close to 0 suggests a weak relation with the independent parameter.
- value close to 1 suggests a strong relation with the independent parameter.

```python

actual_test_values_median_house_values = test_df[['median_house_value']]
actual_test_values_median_income = test_df[['median_income']]
predicted_test_house_values_median_income = lrm_median_income.predict(actual_test_values_median_income)
rsquared_median_income = r2_score(actual_test_values_median_house_values, predicted_test_house_values_median_income)

actual_test_values_population = test_df[['population']]
predicted_test_house_values_population = lrm_population.predict(actual_test_values_population)
rsquared_population = r2_score(actual_test_values_median_house_values, predicted_test_house_values_population)

actual_test_values_households = test_df[['households']]
predicted_test_house_values_households = lrm_households.predict(actual_test_values_households)
rsquared_households = r2_score(actual_test_values_median_house_values, predicted_test_house_values_households)

rsquared = pd.DataFrame({'RMSE': [rsquared_median_income, rsquared_population, rsquared_households]})
```

Combining all of the above calculated values into a dataframe

```python
combined = np.hstack((model_df_test, error_ratio, rmse, rsquared))
summary_df = pd.DataFrame(combined, columns=['Parameter', 'MSE', 'MAE', 'Error Ratio between MAE and Mean', 'RMSE', 'R-Squared'])
display(summary_df)
```
